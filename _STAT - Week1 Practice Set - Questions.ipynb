```python
import math

# Total coins
total_coins = 14

# Odd coins
odd_coins = 5 + 6  # Rs. 1 and Rs. 5 coins

# Total ways to select 3 coins from 14
total_ways = math.comb(total_coins, 3)

# Favorable ways to select 3 odd coins from 11
favorable_ways = math.comb(odd_coins, 3)

# Probability
probability = favorable_ways / total_ways

print("Probability of selecting 3 odd coins:", probability)
```

                </font>

        </td>

</table>

```python
# Write answer here
import math

# Total cards
total_cards = 52

# Total queens
queens = 4

# Total ways to choose 3 cards
total_ways = math.comb(total_cards, 3)

# Favorable ways: 1 queen and 2 non-queens
favorable_ways = math.comb(queens, 1) * math.comb(total_cards - queens, 2)

# Probability
probability = favorable_ways / total_ways

print("Probability of selecting exactly 1 queen:", probability)
```

                    <b>9. The same magician wants another person to select three cards from a pack of 52 cards without replacement. Find the probability that at least one card is a queen?</b>
                </font>
            </div>
        </td>
    </tr>

```python
# Write answer here
import math

# Total cards
total_cards = 52

# Total queens
queens = 4

# Cards selected
cards_selected = 3

# Total ways to select 3 cards
total_ways = math.comb(total_cards, cards_selected)

# Ways to select 3 cards with no queens
ways_no_queen = math.comb(total_cards - queens, cards_selected)

# Probability at least 1 queen
prob_at_least_one_queen = 1 - (ways_no_queen / total_ways)

print("Probability of selecting at least one queen:", prob_at_least_one_queen)
```

```python
# Given probabilities
P_A = 0.2          # Parking in no parking zone
P_B = 0.36         # Day is crowded
P_A_and_B = 0.28   # Parking in no parking zone AND day is crowded

# Conditional probability P(A|B)
P_A_given_B = P_A_and_B / P_B

print("Probability of parking in no parking zone given the day is crowded:", P_A_given_B)
```
