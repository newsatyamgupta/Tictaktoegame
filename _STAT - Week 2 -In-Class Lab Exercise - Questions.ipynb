## 3. Central Limit Theorem

<table align="left">
    <tr>
        <td width="6%">
            <img src="question_icon.png">

        <td>
            <div align="left", style="font-size:120%">
                <font color="#21618C">
                    <b>7. Perform Central Limit Theorem on Left Skewed data with sample size 150 and number of samples as 75.</b>
                </font>
            </div>
        </td>
    </tr>
</table>

```python
# type your code here
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate left-skewed population
# Using Beta distribution (left skewed)
population = np.random.beta(a=5, b=1, size=100000)

# Step 2: Define parameters
sample_size = 150
num_samples = 75

# Step 3: Draw samples and calculate sample means
sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Step 4: Plot original population
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(population, bins=50, kde=True)
plt.title("Left Skewed Population")

# Step 5: Plot sampling distribution of means
plt.subplot(1,2,2)
sns.histplot(sample_means, bins=20, kde=True)
plt.title("Sampling Distribution of Mean (CLT)")

plt.show()
```
